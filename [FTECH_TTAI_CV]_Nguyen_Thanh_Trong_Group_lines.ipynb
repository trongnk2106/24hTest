````python
color_type = {
    "text":  (0, 255,0) ,
     "formula":  (0, 0, 255) ,
     "illustration": (255, 0, 255)
}
````

````python
def get_path(path : str) -> list:
    '''
    Get all path of files in folder
    Args:
        path: path to folder
    Returns:
        list of path
    '''
    return os.listdir(path)
````

````python
def getinfo_anno(path : str) -> dict:
    '''
    Read json file
    Args:
        path: path to json file
    Returns:
        dict of json file

    '''

    with open(path, 'r', encoding='utf-8') as file:
        content = json.load(file)
    return content
````

````python
def calcenter(bbox : tuple) -> tuple:
    '''
    calculate center of bbox
    Args:
        bbox: tuple of bbox
    Returns:
        tuple of center of bbox
    '''
    return (int((bbox[0] + bbox[2]) // 2), int((bbox[1] + bbox[3]) // 2))
````

````python
def statics_height_bbox(bboxes : list) -> int:
    '''
    Statics height of bbox and calculate median height
    Args:
        bboxes: list of bbox
    Returns:
        median height of bbox
    '''

    height = []
    for i in bboxes:
        height.append(i[0][3] - i[0][1])
    height.sort()
    height = height[len(height) // 2]
    return height
````

````python
def grouplines(bboxes : list, width_ : int) -> dict:
    '''
    Group lines by y center of bbox
    Args:
        bboxes: list of bbox
    Returns:
        dict of lines
    '''

    # get median height of bbox
    median_height = statics_height_bbox(bboxes)

    # group lines by y center of bbox
    fairs = {}
    count = 1
    idx = 0
    prev = 0
    while True:

        '''
        Rule base, check type of bbox and distance between 2 center of bbox compared with median height of bbox
        Median height of bbox is calculated by statics_height_bbox function (that statics all of height bbox in image)
        That rule base cover 8 / 10 image, 2 / 10 image wrong because of line is tilted
        '''

        if bboxes[idx][2] !=  'formula' :
            if bboxes[idx + 1][2] == 'text' :
                if bboxes[idx + 1][1][1] - bboxes[idx][1][1] > median_height // 4  :
                    fairs[count] = bboxes[prev:idx + 1]
                    prev = idx + 1
                    count += 1
            elif bboxes[idx + 1][2] == 'formula' :

                if bboxes[idx + 1][1][1] - bboxes[idx][1][1] > median_height // 3  :
                    fairs[count] = bboxes[prev:idx + 1]
                    prev = idx + 1
                    count += 1
            elif bboxes[idx + 1][2] == 'illustration':
                if bboxes[idx + 1][1][1] - bboxes[idx][1][1] > median_height   :
                    fairs[count] = bboxes[prev:idx + 1]
                    prev = idx + 1
                    count += 1
        else:
            if bboxes[idx + 1][1][1] - bboxes[idx][1][1] > median_height //2  :
                fairs[count] = bboxes[prev:idx + 1]
                prev = idx + 1
                count += 1

        idx += 1
        if idx == len(bboxes) - 1:
            fairs[count] = bboxes[prev:]
            break

    # sort by x center of bbox
    for itemkey in fairs:

        fairs[itemkey].sort(key=lambda x: x[1][0])

    return fairs
````

````python
def drawbboxes(anno : dict) -> None:

    '''
    Draw bboxes and put index of line and index of bbox for each bbox
    Args:
        anno: dict of json file
    Returns:
        None
    '''

    start = time.time()

    # read image with opencv
    prefix = './images'
    image =  anno['images'][0]['file_name']

    img = cv2.imread(os.path.join(prefix, image))

    width_, height_ = anno['images'][0]['width'], anno['images'][0]['height']


    # get bbox, and calculate center of bbox for each bbox
    bboxes = [(i['bbox'], calcenter(i['bbox']), i['category'], i['value']) for i in anno['annotations']]
    # sort by y center of bbox
    bboxes.sort(key=lambda x: x[1][1])
    # call function grouplines to group lines by y center of bbox and sort by x center of bbox
    fairs = grouplines(bboxes, width_)

    # if you want to write txt file, you should uncomment the row below
    # file_name_txt = './convert/ ' + image.split('.')[0] + '.txt'

    # write index of line and index of bbox for each bbox to txt file
    # with open(file_name_txt, 'w', encoding='utf-8') as file:
    #     for bbx in fairs:
    #         for ii, i in enumerate(fairs[bbx]):
    #             file.write(f'{i[3]}\t')
    #         file.write('\n')


    # draw bbox and put index of line and index of bbox for each bbox
    for bbx in fairs:

        for ii, i in enumerate(fairs[bbx]):

            xmin = int(i[0][0])
            ymin = int(i[0][1])
            xmax = int(i[0][2])
            ymax = int(i[0][3])

            cv2.putText(img, f'{bbx , ii + 1}', (i[1][0], i[1][1]), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,0,0), 2, cv2.LINE_AA)
            cv2.rectangle(img, (xmin, ymin), (xmax, ymax), color_type[i[2]], 2)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    print(round((time.time() - start), 2))



    plt.imshow(img)
    plt.show()

    # cv2.waitKey(0)
    # cv2.destroyAllWindows()
````

From the given data, after observing the data, I have extracted the following characteristics and developed the algorithm idea for "grouplines" as follows:

First, I calculate the center points of the bounding boxes using the formula x_center = (x_max - x_min) // 2, y_center = (y_max - y_min) // 2.

Then, I sort the bounding boxes in ascending order of their y_center values.

I analyze the heights of the bounding boxes in the image and find the median height of the bounding boxes.

After experimenting and observing the ratio of heights and the difference in center points between the same and different types of bounding boxes, I have created a rule base for the cases designed within the grouplines function. The idea is to use the difference in y-axis values between consecutive bounding box centers. If the difference is smaller than a specified threshold (as represented in the code), it indicates that the two bounding boxes belong to the same line; otherwise, they are considered to be on different lines.

After looping through all the bounding boxes, we can classify them into lines. Then, we continue sorting them based on their x_center values to establish the order of the bounding boxes within the same line.

With the rule base I designed, my algorithm achieves an accuracy of 8/10 images in the dataset. There are discrepancies in 2 images in a few places.
